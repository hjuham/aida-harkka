```python
import pandas as pd
```

```python
df = pd.read_csv("raakaliikennemaara_h_2024-03-05-2024-05-13.csv",delimiter=";")
df['pvm'] = pd.to_datetime(df['pvm'], format='%Y%m%d')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore') 

fig, ax = plt.subplots(figsize=(8, 5))
sns.lineplot(data=df, x='pvm', y='yhteensa', legend='full', ax=ax)
plt.ylim(0,30000)
plt.ylabel("Yhteens")
plt.title("Liikennemrt", fontweight="bold")
plt.grid(True)
plt.show()
```
