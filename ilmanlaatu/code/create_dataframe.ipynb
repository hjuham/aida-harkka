```python
#Dataframen (CSV) luominen Pandas tarkastelua varten. Käytä käsittelemättömiä JSON-tiedostoja GeoPandasilla
import pandas as pd
import datetime
import requests
import os
```

```python
import warnings
warnings.filterwarnings('ignore')
import json

#Yhdistetään kaikki tiedostot samaan dataframeen
df_list = list()
json_dir = '../data'

for filename in os.listdir(json_dir):
    if filename.endswith('.json'):
        try:
            filepath = os.path.join(json_dir, filename)
            # data = json.load(filepath)
            data = pd.read_json(filepath)
            x = pd.json_normalize(data["features"], max_level=10)
            df_list.append(x)
            # print(filepath)
        except:
            print(f"Failed to load {filepath}")

df = pd.concat(df_list).reset_index(drop=True)
#Poistetaan ensimmäinen sarake
df = df.iloc[: , 1:]
df.columns = [col.split(".")[-1] for col in df.columns]
```

```python
df = df.drop_duplicates(subset="id", keep="first")
```

```python
#Dataframe CSV-tiedostoksi
df.to_csv('./data/brushing_data.csv', index=False)
```
